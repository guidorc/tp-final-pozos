```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
######## ANÁLISIS PRODUCCION PETROLEO ########
csv_file = 'produccion/produccion_pet_por_provincia.csv'
df = pd.read_csv(csv_file)

df = df.nsmallest(len(df) - 10, 'avg_pet')

# Plot a boxplot for 'avg_pet' grouped by 'provincia'
plt.figure(figsize=(12, 8))
plt.yticks([n for n in range(0, 1801, 200)])
sns.boxplot(x='provincia', y='avg_pet', data=df, palette='Set3')

# Customize the plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.xlabel('Provincia')
plt.ylabel('Producción')
plt.title('Producción de petróleo por Provincia')

for y in range(0, 1801, 200):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Save the plot as a PNG file and display it
plt.savefig('produccion/boxplot_avg_pet_por_provincia.png', format='png', bbox_inches='tight', dpi=300)
plt.tight_layout()
plt.show()
```

```python
######## ANÁLISIS PRODUCCION GAS ########
csv_file = 'produccion/produccion_gas_por_provincia.csv'
df = pd.read_csv(csv_file)

df = df.nsmallest(len(df) - 10, 'avg_gas')

def quitar_outliers(provincia):
    # Si hay más de 50 datos, borrar los 5 mayores
    if len(provincia) > 50:
      return provincia.nsmallest(len(provincia) - 5, 'avg_gas')
    else:
      return provincia

# Plot a boxplot for 'avg_pet' grouped by 'provincia'
plt.figure(figsize=(12, 8))
plt.yticks([n for n in range(0, 3501, 200)])
sns.boxplot(x='provincia', y='avg_gas', data=df, palette='Set3')

# Customize the plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.xlabel('Provincia')
plt.ylabel('Producción')
plt.title('Producción de gas por Provincia')

for y in range(0, 3501, 200):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Save the plot as a PNG file and display it
plt.savefig('produccion/boxplot_avg_gas_por_provincia.png', format='png', bbox_inches='tight', dpi=300)
plt.tight_layout()
plt.show()
```

```python
######## ANÁLISIS PRODUCCION GAS ########
csv_file = 'produccion/produccion_gas_estado.csv'
df = pd.read_csv(csv_file)

df['avg_gas'] = pd.to_numeric(df['prod_gas'], errors='coerce')

# Plot a boxplot for 'avg_pet' grouped by 'provincia'
plt.figure(figsize=(12, 8))
plt.yticks([n for n in range(0, 85001, 5000)])
sns.boxplot(x='areayacimiento', y='prod_gas', data=df, palette='Set3')

# Customize the plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.xlabel('Yacimiento')
plt.ylabel('Producción')
plt.title('Producción de gas del estado nacional')

for y in range(0, 85001, 5000):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Save the plot as a PNG file and display it
plt.savefig('produccion/boxplot_avg_gas_estado.png', format='png', bbox_inches='tight', dpi=300)
plt.tight_layout()
plt.show()
```
